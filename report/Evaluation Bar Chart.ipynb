```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

topics = ['1 worker','2x1 workers','3x1 workers','2x2 workers',
          '5x1 workers','3x2 workers','7x1 workers','2x4 workers']
fig, ax = plt.subplots(figsize=(20,6))
rects = []
x = np.arange(8)
w = 0.5
q = [21*60+47.025,22*60+39.919,23*60+11.749,24*60+11.151,23*60+56.403,25*60+42.771,26*60+32.473,27*60+30.320]
rects.append(ax.bar(x+0.25, q, w))
    
ax.set_ylabel('Time(s)', fontsize=14)
ax.set_xlabel('Node configuration', fontsize=14)
ax.set_xticks(w/2 + x + 0.25)
ax.set_xticklabels(topics, fontsize=16)

```
